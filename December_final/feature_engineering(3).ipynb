```python
# Check if each value of each column lies within the outlier range.
# If the value is greater than upperOutlier then  True
# If the value is lower than lowerOutlier mark then True
# return the total number of outliers for that column.

```

```python
def numberOfOutliers(mySeries, upperOutlier, lowerOutlier):
    return sum((mySeries > upperOutlier.loc[mySeries.name, ]) |\
               (mySeries < lowerOutlier.loc[mySeries.name,]))

def outliers(df):
    transposedDescribe=(df.describe(include='all').T).round(decimals=3)
    
    transposedDescribe['IQR']=transposedDescribe['75%']-transposedDescribe['25%']
    transposedDescribe['outliers'] = (transposedDescribe['max'] > (transposedDescribe['75%'] + (1.5 * transposedDescribe['IQR'])))\
                            | (transposedDescribe['min'] < (transposedDescribe['25%'] - (1.5 * transposedDescribe['IQR'])))
    IQR = df.quantile(.75) - df.quantile(.25)
    
    upperOutlier = df.quantile(.75) + (1.5 * (IQR))
    lowerOutlier = df.quantile(.25) - (1.5 * (IQR))
    
    transposedDescribe['num_outliers'] = df.apply(numberOfOutliers, args=(upperOutlier, lowerOutlier))
    transposedDescribe.sort_values('num_outliers', ascending=False, inplace=True)
    newColOrder = ['count', 'outliers', 'num_outliers', 'IQR', 'mean', 'std', \
               'min', '25%', '50%', '75%', 'max']
    transposedDescribe = transposedDescribe.reindex(columns=newColOrder)
    
    return transposedDescribe
```
