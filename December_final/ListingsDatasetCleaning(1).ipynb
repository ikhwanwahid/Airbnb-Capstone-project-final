```python
#now we check for missing values in each columns
#get the number and then the % for a clearer pic of missing vs total

missing_values=justlistings_df.isnull().sum().to_frame() #add to frame to keep it in a dataframe
missing_values.columns=['missing']
missing_values['%']=np.round(missing_values['missing']/justlistings_df.shape[0]*100)
missing_values.sort_values(by='missing', ascending=False, inplace=True)
                        
```

```python
#price and extra people are listed as objects , but actually they should be bool
subsetted.dtypes
#need to change them
```

```python
import re #Return the string obtained by replacing the leftmost non-overlapping occurrences of pattern in string by the replacement repl. If the pattern isn’t found
```

```python
subsetted['price'] = subsetted.price.apply(lambda x:re.sub(r'[$,]', '', x)).astype('float')
subsetted['extra_people'] = subsetted.extra_people.apply(lambda x:re.sub(r'[$,]', '', x)).astype('float')
```

```python
#Finally we get dummies all categorical columns!!
category_one_hot_encoding = pd.get_dummies(subsetted[categorical_types])
```

```python
category_one_hot_encoding.head()
```

```python
print(subsetted.shape)
print(category_one_hot_encoding.shape)
```

```python
df_airbnb_clean = pd.concat([subsetted, category_one_hot_encoding],
                            axis=1).drop(categorical_types,
                                                 axis='columns')
df_airbnb_clean.shape
```

```python
df_airbnb_clean.to_csv('airbnb_clean.csv', index=False)
```
