# Gensim LDA Model -Hypertuning with 5 topics

```python
#build our LDA model
def compute_coherence_values(corpus,dictionary,a):
    lda_model = gensim.models.LdaMulticore(corpus=corpus,
                                           id2word=dictionary,
                                           num_topics=5,
                                           workers=3,
                                           random_state=100,
                                           eval_every=1,
                                           chunksize=100,
                                           passes=10,
                                           alpha=a)
    coherence_model_lda=CoherenceModel(model=lda_model,texts=data_lemmatized,dictionary=idtoword,coherence='c_v')
    return coherence_model_lda.get_coherence()

```

```python
import numpy as np
import tqdm
grid = {}
grid['Validation_Set'] = {}
# Topics range
min_topics = 3
max_topics = 20
step_size = 1
topics_range = range(min_topics, max_topics, step_size)
# Alpha parameter
alpha = [0.01, 1, 0.3]
alpha.append('symmetric')
alpha.append('asymmetric')
num_of_docs = len(corpus)
model_results = {
                 'Alpha': [],
                 'Coherence': []
                }
if 1 == 1:
    pbar = tqdm.tqdm(total=540)
    for a in alpha:
        cv = compute_coherence_values(corpus=corpus, dictionary=idtoword,a=a)
        model_results['Alpha'].append(a)
        model_results['Coherence'].append(cv)
        pbar.update(1)
    pd.DataFrame(model_results).to_csv('lda_tuning_results.csv', index=False)
    pbar.close()
```
